```python
import pandas as pd

topic_num = 12
start = 2013 #시작년도
end = 2019 #마지막년도 + 1
dw_start = 1.40015
dw_end = 2.59985
date_range = range(start,end)
'''
1.40015자기상관 X
1.40015 ~ 2.59985 불확실
'''
```

```python
fname = '/content/drive/My Drive/18-2 트랜드/공유자료/파일/sorted_by_date_12topics.p'
documentTopicRate = pd.read_pickle(fname)
documentTopicRate #불러온 dataframe은 다음과 형식을 갖추어야한다.
```

```python
topicRate = pd.DataFrame(columns=range(0,topic_num), index = date_range)#년도별 토픽 비중 평균값을 담을 데이터프레임
```

```python
for date in date_range:
    temp_df = documentTopicRate[(documentTopicRate.index.year == date) ]#특정년도를 뽑아내 임시 데이터프레임에 저장
    for topicType in range(topic_num):
        topicRate.iat[date - start, topicType] = temp_df.iloc[:,topicType].mean()#토픽별로 토픽 비중 평균값을 구함
topicRate
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.stattools import durbin_watson

result = pd.DataFrame(columns=['coef','intersect','p-value', 'Durbin_Waston','Hot/Cold'], index = range(0,topic_num))
result
```

```python
def draw_topic_graph(topicType):
    #input값 설정
    topic = pd.DataFrame({'year': date_range, 'rate':topicRate.iloc[:,topicType]})
    X = topic['year']
    Y = topic['rate']
    
    #선형 회귀 그래프 그리기 by seaborn
    print(sns.regplot(x=X, y=Y))
    
    #선형 회귀 적용 by statsmodel
    X = sm.add_constant(X)
    model = sm.OLS(Y.astype(float),X).fit()
    result.iat[topicType,0] = model.params[1]
    result.iat[topicType,1] = model.params[0]
    result.iat[topicType,2] = model.pvalues[0]
    result.iat[topicType,3] = durbin_watson(model.wresid)
    isHotCold(topicType, result)
    
    #선형 회귀 그래프 그리기 by statsmodel
    print("Topic",topicType)
    print(model.summary())
    
    x = [a for a in  date_range]
    y = [model.params[1]*b+model.params[0] for b in date_range]
    plt.plot(x,y) 
    plt.scatter(topic.year, topic.rate)
    plt.xlabel('YEAR')
    plt.ylabel('RATE')
    plt.title('Topic'+str(topicType)+ ' Graph')
    plt.show()
    
def isHotCold(topicType, result):#
    coef = result.iat[topicType,0]
    p_value = result.iat[topicType,2]
    dw = result.iat[topicType,3]
    
    #Hot/Cold인지 구분하는 부분
    if ( p_value < 0.05) & (dw > dw_start) & (dw < dw_end):
        if coef > 0: result.iat[topicType,4] = 'Hot'
        else: result.iat[topicType,4] = 'Cold'
    else:#hot/cold를 말할 수 없는 토픽들
        result.iat[topicType,4] = '-'
```

```python
for i in range(topic_num):
    draw_topic_graph(i)
```
